# **Findings:**
1. TV Advertising vs. Newspaper Advertising:

The analysis revealed a strong positive correlation between TV advertising and sales performance. Companies that allocated a higher budget to TV advertising saw a corresponding increase in sales.
Newspaper advertising also had a positive impact on sales, but the correlation was weaker than that of TV advertising. Companies investing in newspaper ads experienced a moderate increase in sales.


2. TV Advertising vs. Radio Advertising:

The correlation between TV advertising and sales performance was more pronounced than that between Radio advertising and sales. Companies that advertised on TV observed a more substantial increase in sales compared to those who opted for Radio.
Radio advertising, while still contributing to sales growth, was not as effective as TV advertising in driving revenue.

# Define Features and Labels

We wil predict which medium be more suitable for more purchases to increase Sales

```python
x=df[["TV"]]
y=df["Sales"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
model.predict([[44.5]])
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
